# Recommendation Algorithm

```python
def RecSys(inpid):
    cid = df["customer_id"].tolist()
    cid = Remove(cid)
    #cid.remove(inpid)
    userRatings = df[df['customer_id'] == inpid]
    print("Reading user ratings...")
    loved = userRatings[userRatings['star_rating'] >= 4]
    hated = userRatings[userRatings['star_rating'] <= 3]
    print("Loved Products of "+str(inpid))
    print(loved)
    print("Hated Products of "+str(inpid))
    print(hated)
    recitem = []
    print("Training the System...")
    for i in cid:
        tempt = df[df['customer_id'] == i]
        temlis = tempt['product_parent'].tolist()
        for item in loved['product_parent']:
            if item in temlis:
                indexit = temlis.index(item)
                if len(temlis) > indexit+1 :
                    print(temlis[indexit:])
                    recitem = recitem + temlis[indexit:]
    recitemsorder = [item for items, c in Counter(recitem).most_common() for item in [items]] 
    itemstorec = recitemsorder
    for item in itemstorec:
        if item in hated:
            del itemstorec[itemstorec.index(item)]
    finitems = []
    for num in itemstorec:
        finitems = finitems + df[df['product_parent'] == num]['product_title'].tolist()
    finitems = Remove(finitems)
    print(finitems)
    return finitems
```

```python
uids = df['customer_id']
uids
```

```python
from collections import Counter 
  
def most_frequent(List): 
    occurence_count = Counter(List) 
    return occurence_count.most_common(1)[0][0]
def Remove(duplicate): 
    return list(dict.fromkeys(duplicate))
```

```python
print("Most number of Purchases done by: ",most_frequent(uids))
```

```python
inid = int(input("Enter a User ID: "))
```

```python
outputrec = RecSys(inid)
print("Output Recommendations")
outputrec
```
